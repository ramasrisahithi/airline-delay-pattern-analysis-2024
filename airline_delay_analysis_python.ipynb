# **Data Cleaning**

```python
#Checking missing values
missing_count = df.isnull().sum()
missing_pct   = (df.isnull().sum() / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count' : missing_count,
    'Missing %'     : missing_pct
})

missing_df[missing_df['Missing Count'] > 0]
```

```python
#Check Duplicate rows
print("Duplicate rows :", df.duplicated().sum())
```

```python
#Fix Date Column to Proper Format
df['fl_date'] = pd.to_datetime(df['fl_date'])
df['fl_date'].dtype
```

```python
#Fixing Missing Values in Delay Columns
# Define delay columns first
delay_cols = ['carrier_delay', 'weather_delay', 'nas_delay',
              'security_delay', 'late_aircraft_delay']

# Fill missing values with 0
df[delay_cols] = df[delay_cols].fillna(0)

# Verify no nulls remain
df[delay_cols].isnull().sum()
```

```python
#Remove Bad rows
# Remove rows where arrival delay is missing and flight is not cancelled
before = len(df)

df = df[~((df['arr_delay'].isnull()) & (df['cancelled'] == 0))]

after = len(df)
print("Rows removed   :", before - after)
print("Rows remaining :", after)
```

# **Feature Engineering**

```python
#Extract Month Name
# month column already exists — just add month name
df['month_name'] = pd.to_datetime(df['fl_date']).dt.month_name()
df['month_name'].head()
```

```python
# Extract Day of Week Name
# day_of_week already exists as number — add day name
df['day_name'] = pd.to_datetime(df['fl_date']).dt.day_name()
df['day_name'].head()
```

```python
#Extract Quarter
df['quarter'] = pd.to_datetime(df['fl_date']).dt.quarter
df['quarter'].head()
```

```python
# Create Total Delay Column
df['total_delay'] = df[delay_cols].sum(axis=1)
df[['total_delay'] + delay_cols].head()
```

```python
#Create Is Delayed Column
# BTS official definition — more than 15 min arrival delay = delayed
df['is_delayed'] = (df['arr_delay'] > 15).astype(int)
df['is_delayed'].value_counts()
```

```python
#Create Delay Severity Column
bins   = [-999, 0, 15, 30, 60, 999]
labels = ['Early/On-time', 'Minor (0-15)',
          'Moderate (15-30)', 'Severe (30-60)', 'Critical (60+)']
df['delay_severity'] = pd.cut(df['arr_delay'].fillna(0), bins=bins, labels=labels)
df['delay_severity'].value_counts()
```

```python
#Create Route Column
df['route'] = df['origin'] + ' to ' + df['dest']
df['route'].head()
```

```python
# Preview All New Columns Added
df[['fl_date', 'month', 'month_name', 'day_name',
    'quarter', 'total_delay', 'is_delayed',
    'delay_severity', 'route']].head()
```

# **Analysis & Visualizations**

```python
#import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for all charts
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
#Airline Performance (On-Time % by Airline)
airline_perf = df.groupby('op_unique_carrier').agg(
    total_flights  = ('fl_date', 'count'),
    delayed_flights = ('is_delayed', 'sum'),
    avg_delay      = ('arr_delay', 'mean')
).reset_index()

airline_perf['on_time_pct'] = ((1 - airline_perf['delayed_flights'] / airline_perf['total_flights']) * 100).round(2)
airline_perf = airline_perf.sort_values('on_time_pct', ascending=False)

print(airline_perf)
```

```python
#Plot On-Time % by Airline
plt.figure(figsize=(12, 5))
sns.barplot(data=airline_perf, x='op_unique_carrier', y='on_time_pct', palette='Blues_d')
plt.title('On-Time Performance by Airline (2024)')
plt.xlabel('Airline')
plt.ylabel('On-Time %')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
#Monthly Delay Trend
monthly = df.groupby('month').agg(
    avg_delay      = ('arr_delay', 'mean'),
    total_flights  = ('fl_date', 'count'),
    delayed_flights = ('is_delayed', 'sum')
).reset_index()

monthly['on_time_pct'] = ((1 - monthly['delayed_flights'] / monthly['total_flights']) * 100).round(2)

print(monthly)
```

```python
#Plot Monthly Delay Trend
plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly, x='month', y='avg_delay', marker='o', color='steelblue')
plt.title('Average Arrival Delay by Month (2024)')
plt.xlabel('Month')
plt.ylabel('Avg Delay (mins)')
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun',
                           'Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.show()
```

```python
#Delay Cause Breakdown
delay_cols = ['carrier_delay', 'weather_delay', 'nas_delay',
              'security_delay', 'late_aircraft_delay']

cause_totals = df[delay_cols].sum()
cause_pct    = (cause_totals / cause_totals.sum() * 100).round(2)

print(cause_pct)
```

```python
#Plot Delay Cause Breakdown
plt.figure(figsize=(8, 8))
plt.pie(cause_totals,
        labels=['Carrier', 'Weather', 'NAS', 'Security', 'Late Aircraft'],
        autopct='%1.1f%%',
        colors=['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2'])
plt.title('Delay Cause Breakdown (2024)')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Most Delayed Routes
route_delay = df.groupby('route').agg(
    avg_delay     = ('arr_delay', 'mean'),
    total_flights = ('fl_date', 'count')
).reset_index()

# Only routes with minimum 100 flights
route_delay = route_delay[route_delay['total_flights'] >= 100]
top10_routes = route_delay.nlargest(10, 'avg_delay')

print(top10_routes)
```

```python
# Plot Top 10 Most Delayed Routes
plt.figure(figsize=(12, 5))
sns.barplot(data=top10_routes, x='avg_delay', y='route', palette='Reds_d')
plt.title('Top 10 Most Delayed Routes (2024)')
plt.xlabel('Avg Delay (mins)')
plt.ylabel('Route')
plt.tight_layout()
plt.show()
```

```python
# Delay by Day of Week
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

day_delay = df.groupby('day_name')['arr_delay'].mean().reindex(day_order).reset_index()
day_delay.columns = ['day_name', 'avg_delay']

print(day_delay)
```

```python
# Plot Delay by Day of Week
plt.figure(figsize=(10, 5))
sns.barplot(data=day_delay, x='day_name', y='avg_delay', palette='coolwarm')
plt.title('Average Delay by Day of Week (2024)')
plt.xlabel('Day')
plt.ylabel('Avg Delay (mins)')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Most Delayed Airports
airport_delay = df.groupby('origin').agg(
    avg_delay     = ('arr_delay', 'mean'),
    total_flights = ('fl_date', 'count')
).reset_index()

airport_delay = airport_delay[airport_delay['total_flights'] >= 500]
top10_airports = airport_delay.nlargest(10, 'avg_delay')

print(top10_airports)
```

```python
#Plot Top 10 Most Delayed Airports
plt.figure(figsize=(12, 5))
sns.barplot(data=top10_airports, x='avg_delay', y='origin', palette='Oranges_d')
plt.title('Top 10 Most Delayed Airports (2024)')
plt.xlabel('Avg Delay (mins)')
plt.ylabel('Airport Code')
plt.tight_layout()
plt.show()
```
